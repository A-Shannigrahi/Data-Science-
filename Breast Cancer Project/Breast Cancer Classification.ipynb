# Model Building 

```python
x=df_cancer.iloc[:,:-1].values
y=df_cancer.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
```

# Testing the Model 

```python
y_test_predicted=log_reg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
matrix=pd.DataFrame(confusion_matrix(y_test,y_test_predicted))
matrix.index=["Predicted 0","Predicted 1"]
matrix.columns=["Actual 0","Actual 1"]
print(matrix)
```

```python
print(accuracy_score(y_test,y_test_predicted))
```

# 96.49 % Accuracy Obtained Using Logistic Regression Model
